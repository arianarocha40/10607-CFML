```python
class Stack:
    def __init__(this):
        this.elements = []

    def push(this, item):
        this.elements.append(item)

    def pop(this):
        if not this.is_empty():
            return this.elements.pop()
        else:
            raise IndexError("pop from empty stack")

    def is_empty(this):
        return len(this.elements) == 0


class Queue:
    def __init__(this):
        this.contents = []

    def enqueue(this, element):
        this.contents.append(element)

    def dequeue(this):
        if not this.is_empty():
            return this.contents.pop(0)
        else:
            raise IndexError("dequeue from empty queue")

    def is_empty(this):
        return len(this.contents) == 0


class TreeNode:
    def __init__(this, attribute, end=False, outcome=None):
        this.attribute = attribute
        this.kids = []
        this.end = end
        this.outcome = outcome

    def add_kid(this, kid):
        this.kids.append(kid)


class DecisionTree:
    def __init__(this):
        this.base_vertex = None

    def set_root(this, base):
        this.base_vertex = base

    def golf_tree_dfs(this):
        path_stack = Stack()
        path_stack.push((this.base_vertex, []))
        rules_collector = []

        while not path_stack.is_empty():
            current_kid, rule_path = path_stack.pop()

            if current_kid.end:
                complete_rule = " and ".join(rule_path)
                rules_collector.append(f"If {complete_rule}, Golf={current_kid.outcome}")
            else:
                for sub_kid in reversed(current_kid.kids):
                    new_path = rule_path + [f"{current_kid.attribute}={sub_kid.attribute}"]
                    path_stack.push((sub_kid, new_path))

        for rule in sorted(rules_collector):
            print(rule)

    def golf_tree_bfs(this):
        path_queue = Queue()
        path_queue.enqueue((this.base_vertex, []))
        rules_collector = []

        while not path_queue.is_empty():
            current_kid, rule_path = path_queue.dequeue()

            if current_kid.end:
                complete_rule = " and ".join(rule_path)
                rules_collector.append(f"If {complete_rule}, Golf={current_kid.outcome}")
            else:
                for sub_kid in current_kid.kids:
                    new_path = rule_path + [f"{current_kid.attribute}={sub_kid.attribute}"]
                    path_queue.enqueue((sub_kid, new_path))

        for rule in sorted(rules_collector):
            print(rule)


if __name__ == "__main__":
    base_vertex = TreeNode("Outlook")

    overcast_kid = TreeNode("overcast", end=True, outcome="yes")
    rain_kid = TreeNode("rain")
    sunny_kid = TreeNode("sunny")

    wind_kid = TreeNode("Wind")
    wind_kid.add_kid(TreeNode("false", end=True, outcome="yes"))
    wind_kid.add_kid(TreeNode("true", end=True, outcome="no"))
    rain_kid.add_kid(wind_kid)

    humidity_kid = TreeNode("Humidity")
    humidity_kid.add_kid(TreeNode("high", end=True, outcome="no"))
    humidity_kid.add_kid(TreeNode("low", end=True, outcome="yes"))
    sunny_kid.add_kid(humidity_kid)

    base_vertex.add_kid(rain_kid)
    base_vertex.add_kid(sunny_kid)
    base_vertex.add_kid(overcast_kid)

    golf_tree = DecisionTree()
    golf_tree.set_root(base_vertex)

    print("Depth First Search:")
    golf_tree.golf_tree_dfs()

    print("\nBreadth First Search:")
    golf_tree.golf_tree_bfs()
```
